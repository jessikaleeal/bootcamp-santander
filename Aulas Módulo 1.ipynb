#Estruturas Condicionais
Desviar ações do algoritmo dependendo de algumas ações

```python
valor_passagem = 4.30
valor_corrida = input('Qual é o valor da corrida? ')

if float(valor_corrida) <= valor_passagem * 5:
  print('Pague a corrida!')
if float(valor_corrida) > valor_passagem * 5:
  print("Pegue o ônibus")
```

```python
#simplificando o código

valor_passagem = 4.30
valor_corrida = input('Qual é o valor da corrida? ')

if float(valor_corrida) <= valor_passagem * 5:
  print('Pague a corrida!')
else:
  print("Pegue o ônibus")
```

```python
#Criando mais um condicional

valor_passagem = 4.30
valor_corrida = input('Qual é o valor da corrida? ')

if float(valor_corrida) <= valor_passagem * 5:
  print('Pague a corrida!')
else:
  if float(valor_corrida) <= valor_passagem * 6:
    print("Aguarde um momento, o valor por abaixar")
  else:
    print("Pegue o ônibus")
```

#Estruturas de **Repetição - While**
Pedir que uma determinada tarefa seja feita enquanto uma condição é verdadeira

```python
contador = 0

while contador < 10:
  contador = contador+1
  if contador == 1:
    print(contador, 'Item limpo')
  else:
    print(contador, 'Itens limpos')

print('Fim da repetição do bloco while')
```

```python
#break não é uma boa prática

contador = 0

while True:
  if contador < 10:
      contador = contador+1
      if contador == 1:
        print(contador, 'Item limpo')
      else:
        print(contador, 'Itens limpos')
  else:
    break

print('Fim da repetição do bloco while')
```

```python
texto = input('Digite a sua senha: ')

while texto != 'LetsCode':
  texto = input('Senha inválida. Tente novamente')

print('Acesso permitido')
```

```python
contador = 0

while contador < 10:
  contador = contador+1
  if contador == 1:
    continue
  print(contador, 'Itens limpos')


print('Fim da repetição do bloco while')
```
