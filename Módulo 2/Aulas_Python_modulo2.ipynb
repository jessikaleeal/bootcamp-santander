# **Aula 1 - Tuplas**
Diferença entre a lista e a tupla é flexibilidade

Lista: inserir, remover, alterar

Tupla: não é possível mexer nos elementos dentro dela // Menor em questão de armazenamento


```python
nomes_paises = 'Brasil', 'Argentina', 'China', 'Canadá', 'Japão'
# nomes_paises = ('Brasil', 'Argentina', 'China', 'Canadá', 'Japão')

type(nomes_paises)
```

```python
nome_estado = 'São Paulo',
print(nome_estado)
type(nome_estado)
```

```python
nomes_paises[0]
```

```python
# operação que faz na tupla que a lista não faz
# pegar todos os valores que estão juntos em uma tupla só e colocar cada um em uma variável diferente
# função de ampec

b, a, c, ca, j = nomes_paises
print(b, c, j)
```

```python
# ao invés de imprimir a tupla inteira, imprimir só os valores
print(*nomes_paises)
```

```python

#pedindo pro usuário inserir o valor
nome_cidade = input('Que cidade do Brasil é conhecida como cidade maravilhosa? ')

# tira qualquer espaço que o usuário tenha colocado
nome_cidade = nome_cidade.strip()

# enquanto o nome da cidade digitado for diferente de rio de janeiro, pede pra ele colocar de novo o resultado
# utilizar o tratamento de lower para deixar tudo em maiusculo o que a pessoa colocar
# entrada de valor textual e um valor definido
while nome_cidade.lower() != 'rio de janeiro':
  print('Tenta de novo, vai')
  nome_cidade = input('Que cidade do Brasil é conhecida como cidade maravilhosa? ')

print('Booa, campeão!')
```

#**Aula 5 - Estruturas de repetição**
for -> navegar por um objeto / utilizado em um objeto que precisa ser navegado em todos os elementos / precisa ser repetido em um determinado período de vezes

while -> sequencia de tarefas executadas enquanto uma condição é verdadeira

```python
nomes_cidades = ['São Paulo', 'Londres', 'Tóquio', 'Paris']

#nome representa os itens dentro da lista
for nome in nomes_cidades:
  print(nome)
```

```python
# mesmo exemplo utilizando o while 

contador = 0
nomes_cidades = ['São Paulo', 'Londres', 'Tóquio', 'Paris', 'Roma']
while contador < len(nomes_cidades):
  print(nomes_cidades[contador])
  contador = contador + 1
```

```python
nomes_cidades = 'São Paulo', 'Londres', 'Tóquio', 'Paris'

#nome representa os itens dentro da lista
for nome in nomes_cidades:
  print(nome)
```

```python
# usando com um dicionário

cidade = {
    'nome': 'São Paulo',
    'estado': 'São Paulo',
    'populacao_milhoes': 12.2
}
for chave in cidade:
  print(f' {chave}: {cidade[chave]}')
```

```python
nomes_cidades = 'São Paulo', 'Londres', 'Tóquio', 'Paris'
for nome in nomes_cidades:
  # não muda o elemento pois é uma cópia
  nome = 'Rio de Janeiro'
print(nomes_cidades)
```

```python
for posicao in range(len(nomes_cidades)):
  print(posicao)
```

#**Aula 07 - Funções II**

```python
def calcula_media(valor1, valor2,valor3):
  soma = valor1 + valor2 + valor3
  media = soma / 3
  return media
```

```python
def calcula_media(*args):
  print(args, type(args))
```

```python
calcula_media(10,8,9)
```

```python

#media com margem de erro

def calcula_media(*args, margem):
  soma = sum(args)
  media = soma / len(args)
  return media + margem
```

```python
calcula_media(10,8,9, margem = 0.3)
```

```python
# tecnica utilizada para ação padrão com parâmetros definidos

def print_info(**kwargs):
  print(kwargs, type(kwargs))
```
