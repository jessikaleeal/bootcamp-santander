# Aula 1 - Manipulação de arquivos

```python
# 'R' -> abrindo em modo de leitura
arquivo = open('dom_casmurro_cap_1.txt.txt', 'r')

# abrindo o arquivo
texto = arquivo.read()
print(texto)
# fechando o arquivo
arquivo.close()
```

```python
# analisando linha a linha do arquivo
arquivo = open('dom_casmurro_cap_1.txt.txt', 'r')

# só acessa uma linha do arquivo
linha = arquivo.readline()
while linha != '':
  print(linha, end = '')
  linha = arquivo.readline()

arquivo.close()
```

```python
# utilizando o for ao invés do while

arquivo = open('dom_casmurro_cap_1.txt.txt', 'r')
for linha in arquivo:
  print(linha, end ='')
arquivo.close()
```

```python
# utilizando with para fechar o arquivo ao invés de fechar no final

with open('dom_casmurro_cap_1.txt.txt', 'r') as arquivo:
  texto = arquivo.read()
  print(texto)
```

```python
with open('dom_casmurro_cap_1.txt.txt', 'w') as arquivo:
  arquivo.write("Essa é uma linha que eu escrevi usando Python\n")
  arquivo.write("Essa é a segunda linha que eu escrevi usando Python\n")
```

```python
with open('dom_casmurro_cap_1.txt.txt', 'r') as arquivo:
  print(arquivo.read(), end = '')
```

# AULA 2 - CSV

```python
import csv
```

```python
with open('brasil_covid.csv', 'r') as arquivo_csv:
  leitor = csv.reader(arquivo_csv)
  # ignorando o cabeçalho
  header = next(leitor)
  for linha in leitor:
    # mostrar os registros que apenas novos casos são maiores que 1
    for linha in leitor:
      if float(linha[2]) > 1:
          print(linha)
```

```python
# mesmo resultado acima sem utilizar a biclioteca
# leitura de um arquivo CSV python

with open('brasil_covid.csv') as csv_file:
  linhas = csv_file.read()
  linhas = linhas.split('\n')
  for linha in linhas:
      linha = linha.split(',')
      print(linha)
```

```python
# escrevendo um arquivo CSV

with open('users.csv', 'w') as arquivo_users:
  escritor = csv.writer(arquivo_users)
  escritor.writerow(['nome', 'sobrenome', 'email', 'genero'])
  escritor.writerow(['Pietro', 'Ribeiro', 'pietro@gmail.com', 'Masculino'])
  
```
